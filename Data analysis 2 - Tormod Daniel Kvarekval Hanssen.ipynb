```python
import math

group_yes = df2[df2["paid"] == "yes"]["Scores"]
group_no = df2[df2["paid"] == "no"]["Scores"]

mean_yes = group_yes.mean()
mean_no = group_no.mean()

std_yes = group_yes.std(ddof=1)
std_no = group_no.std(ddof=1)

n_yes = len(group_yes)
n_no = len(group_no)
```

```python
t = (mean_yes - mean_no) / math.sqrt(
    (std_yes**2 / n_yes) + (std_no**2 / n_no)
)

print("t-stat:", t)
```

```python
de_f = n_yes + n_no - 2
print("degree of freedom:", de_f)
```

```python
from scipy.stats import ttest_ind

group_yes = df2[df2["paid"] == "yes"]["Scores"]
group_no = df2[df2["paid"] == "no"]["Scores"]

mean_yes = group_yes.mean()
mean_no = group_no.mean()

t_stat, p_val = ttest_ind(group_yes, group_no, equal_var=False)

print(f"Mean (paid=yes): {mean_yes:.2f}")
print(f"Mean (paid=no): {mean_no:.2f}")
print(f"Difference: {mean_yes - mean_no:.2f}")
print(f"t-statistic: {t_stat:.3f}")
print(f"p-value: {p_val:.4f}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

target = "Scores"
y = df2[target]
```

```python
features_base = ["failures"]

features_core = ["failures", "studytime", "parent_edu_avg"]

features_extended = [
    "failures",
    "studytime",
    "parent_edu_avg",
    "goout",
    "absences",
    "health"
]

X_base = df2[features_base]
X_core = df2[features_core]
X_extended = df2[features_extended]

# Train multiple linear regression models (baseline, core, extended)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_core, y, test_size=0.2, random_state=42
)
```

```python
# Base
X_train_base = X_base.loc[X_train.index]
X_test_base = X_base.loc[X_test.index]

# Core 
X_train_core = X_train
X_test_core = X_test

# Extended
X_train_ext = X_extended.loc[X_train.index]
X_test_ext = X_extended.loc[X_test.index]
```

```python
# Baseline model
model_base = LinearRegression()
model_base.fit(X_train_base, y_train)

y_pred_base = model_base.predict(X_test_base)

rmse_base = np.sqrt(mean_squared_error(y_test, y_pred_base))
r2_base = r2_score(y_test, y_pred_base)


# Core model
model_core = LinearRegression()
model_core.fit(X_train_core, y_train)

y_pred_core = model_core.predict(X_test_core)

rmse_core = np.sqrt(mean_squared_error(y_test, y_pred_core))
r2_core = r2_score(y_test, y_pred_core)


# Extended model
model_ext = LinearRegression()
model_ext.fit(X_train_ext, y_train)

y_pred_ext = model_ext.predict(X_test_ext)

rmse_ext = np.sqrt(mean_squared_error(y_test, y_pred_ext))
r2_ext = r2_score(y_test, y_pred_ext)
```

```python
print("Baseline:")
print(f"RMSE: {rmse_base:.2f}, R²: {r2_base:.2f}")

print("\nCore:")
print(f"RMSE: {rmse_core:.2f}, R²: {r2_core:.2f}")

print("\nExtended:")
print(f"RMSE: {rmse_ext:.2f}, R²: {r2_ext:.2f}")

# Compare performance using RMSE and R²
```

```python
plt.figure(figsize=(15,5))

# Baseline
plt.subplot(1,3,1)
plt.scatter(y_test, y_pred_base)
plt.plot([0,20], [0,20], linestyle='-', color="r")
plt.title(f"Baseline\nRMSE={rmse_base:.2f}, R²={r2_base:.2f}")
plt.xlabel("Actual Score")
plt.ylabel("Predicted Score")

# Core 
plt.subplot(1,3,2)
plt.scatter(y_test, y_pred_core)
plt.plot([0,20], [0,20], linestyle='-', color="m")
plt.title(f"Core\nRMSE={rmse_core:.2f}, R²={r2_core:.2f}")
plt.xlabel("Actual Score")

# Extended
plt.subplot(1,3,3)
plt.scatter(y_test, y_pred_ext)
plt.plot([0,20], [0,20], linestyle='-', color="g")
plt.title(f"Extended\nRMSE={rmse_ext:.2f}, R²={r2_ext:.2f}")
plt.xlabel("Actual Score")

plt.tight_layout()
plt.show()
```

```python
def map_binary_columns(df):
    df_copy = df.copy()
    
    for col in df_copy.select_dtypes(include="object").columns:
        unique_vals = df_copy[col].unique()
        
        if set(unique_vals) <= {"yes", "no"}:
            df_copy[col] = df_copy[col].map({"no": 0, "yes": 1})
            
        elif set(unique_vals) <= {"F", "M"}:
            df_copy[col] = df_copy[col].map({"F": 0, "M": 1})
    
    return df_copy
```

```python
# Map simple binary columns (yes/no, F/M) to 0/1
X_processed = map_binary_columns(X)
```
